```python
from langgraph.graph import StateGraph,START,END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
#state
class BlogState(TypedDict):
    title : str
    outline : str
    content : str
    evaluate : int
```

```python
def create_outline(state : BlogState) -> BlogState:
    title = state['title']
    prompt = f"create a details outline of the topic {title}"
    state["outline"] = model.invoke(prompt).content

    return state
```

```python
def create_blog(state : BlogState) -> BlogState:
    title = state["title"]
    outline = state["outline"]
    prompt = f"create a 200 word blog of the title - {title} and the outline of the title \n {outline}"
    state["content"] = model.invoke(prompt).content

    return state
```

```python
def evaluate_score(state : BlogState) -> BlogState:
    title = state["title"]
    outline = state["outline"]

    prompt = f"Based of the title - {title} \n score my outline - {outline}"
    state["evaluate"] = model.invoke(prompt).content

    return state
```

```python
graph = StateGraph(BlogState)
graph.add_node("create_outline",create_outline)
graph.add_node("create_blog",create_blog)
graph.add_node("evaluate_score",evaluate_score)


graph.add_edge(START,"create_outline")
graph.add_edge("create_outline","create_blog")
graph.add_edge("create_blog","evaluate_score")
graph.add_edge("evaluate_score",END)
```

```python
workflow = graph.compile()
```

```python
initial_state = {"title":"Rise of AI in india"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
