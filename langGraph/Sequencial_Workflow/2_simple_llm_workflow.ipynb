```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash")
```

```python
#define state
class llmState(TypedDict):
    question : str
    answer : str
```

```python
def llmqa(state : llmState) -> llmState:
    #extract question from graph
    question = state["question"]
    #form a prompt
    prompt= f"generate the answer of the question \n {question}"
    #get a response form llm
    response = llm.invoke(prompt)
    #update state
    state["answer"] = response

    return state
```

```python
#define graph
graph = StateGraph(llmState)

#define nodes
graph.add_node("llmqa",llmqa)

#define edges
graph.add_edge(START,"llmqa")
graph.add_edge("llmqa",END)

#compile graph
workflow = graph.compile()
```

```python
#execute graph
initial_state = {"question":"who is the prime minister of india"}
final_state = workflow.invoke(initial_state)
print(final_state["answer"].content)
```
